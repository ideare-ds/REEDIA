# AHP: valutazione dei progetti R&D

## Definizioni

$C$ matrice dei confronti a coppie tra i criteri
\begin{equation*}
    C = \begin{pmatrix}
     c_{11} &c_{12} &c_{13} &c_{14}  \\ 
     c_{21} &c_{22} &c_{23} &c_{24}  \\ 
     c_{31} &c_{32} &c_{33} &c_{34}  \\ 
     c_{41} &c_{42} &c_{43} &c_{44}  
    \end{pmatrix}
\end{equation*}

$A^{(c)}$ matrice dei confronti a coppie tra i sottocriteri del criterio $c$
\begin{equation*}
    A^{(c)} = \left(a^{(c)}_{ij}\right) = \begin{pmatrix}
     a^{(c)}_{11} & \cdots  & a^{(c)}_{1m_c} \\ 
     \vdots & \ddots & \vdots \\
    a^{(c)}_{m_c1} & \cdots & a^{(c)}_{m_cm_c}\end{pmatrix}
\end{equation*}

$v$ vettore di priorità locale dei criteri
\begin{equation*} 
   v^*=\begin{pmatrix}
   \sqrt[4]{c_{11}\,c_{12}\,c_{13}\,c_{14}\,} \\ 
   \sqrt[4]{c_{21}\,c_{22}\,c_{23}\,c_{24}\, }\\ 
   \sqrt[4]{c_{31}\,c_{32}\,c_{33}\,c_{34}\, }\\ 
   \sqrt[4]{c_{41}\,c_{42}\,c_{43}\,c_{44}\, }\\ 
   \end{pmatrix}
\end{equation*}

\begin{equation*} 
   v=\frac{v^*}{\sum_{i=1}^4 v_i^*}
\end{equation*}

$w^{(c)}$ vettore di priorità locale dei sottocriteri del criterio $c$
\begin{equation*} 
   w^{(c)*}=\begin{pmatrix}
    \sqrt[m_c]{a_{11}\,a_{12}\, \cdots \,a_{1m_c}} \\ 
    \sqrt[m_c]{a_{21}\,a_{22}\, \cdots \,a_{2m_c}}\\ 
    \vdots \\ 
    \sqrt[m_c]{a_{m_c 1}\,a_{m_c 2}\, \cdots \,a_{m_c m_c}}\\ 
   \end{pmatrix}
\end{equation*}

\begin{equation*} 
   w^{(c)}= \frac{w^{(c)*}}{\sum_{i=1}^{m_c} w_i^{(c)*}}
\end{equation*}

W è una matrice a blocchi $4\times N_{ind}$ con $N_{ind} = \sum_c m_c$ numero totale di indicatori 
\begin{equation*} 
   W=\begin{pmatrix}
     w^{(1)\, T} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt}  & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} \\ 
    \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & w^{(2)\, T} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} \\
    \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & w^{(3)\, T} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} \\
    \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & w^{(4)\, T} 
   \end{pmatrix}
\end{equation*}

peso globale dell'indicatore $i$ è dato da
\begin{equation*} 
    P_i = (v^T W)_i = \sum_j v_{j} \, W_{ji}
\end{equation*}

Per ogni decisore $k$ avremo un vettore di pesi globali $P^{(k)}$. 
L'aggregazione collettiva  è data da
\begin{equation*}
    P^{(group)}_i = \frac{\prod_{k=1}^{n_k} \left(P_i^{(k)}\right)^{\frac{1}{n_k}}}{\sum_{j=1}^{N_{ind}}\prod_{k=1}^{n_k} \left(P_j^{(k)}\right)^{\frac{1}{n_k}}} 
    =\frac{1}{1+\frac{\sum_{j\neq i}\prod_{k=1}^{n_k} \left(P_j^{(k)}\right)^{\frac{1}{n_k}}}{\prod_{k=1}^{n_k} \left(P_i^{(k)}\right)^{\frac{1}{n_k}}}}
\end{equation*}

\begin{equation*} 
   S_{DPR\&DI} = P^{(group)} x = \sum_{i=1}^{N_{ind}} P^{(group)}_i x_i 
\end{equation*}

\begin{equation*} 
    \sigma^{2} = \frac{1}{3}\sum_{i=1}^{3}\sum_{j=i+1}^{4}\left[\log\left(\frac{c_{ij}}{v_i/v_j}\right)\right]^2
\end{equation*}
\begin{equation*} 
    {\sigma^{(c)}}^2 = \frac{2}{(m_c-1)(m_c-2)}\sum_{i=1}^{m_c-1}\sum_{j=i+1}^{m_c}\left[\log\left(\frac{a_{ij}^{(c)}}{w_i^{(c)}/w_j^{(c)}}\right)\right]^2
\end{equation*}

\begin{equation*} 
    \sigma_{v_i}^{2} = \frac{15}{16}\left[\sum_{i=1}^{4} v_i^2 -v_i^2\right]\sigma^2 v_i^2
\end{equation*}
\begin{equation*} 
    \sigma_{w_i^{(c)}}^{2} = \frac{m_c^2-1}{m_c^2}\left[\sum_{i=1}^{m_c} {w_i^{(c)}}^2 -{w_i^{(c)}}^2\right]{\sigma^{(c)}}^2 w_i^2
\end{equation*}

\begin{equation*} 
   \sigma_W^2=\begin{pmatrix}
     \left(\sigma_{w^{(1)}}^2\right)^T & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt}  & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} \\ 
    \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \left(\sigma_{w^{(2)}}^2\right)^T & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} \\
    \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \left(\sigma_{w^{(3)}}^2\right)^T & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} \\
    \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \rule[.5ex]{1em}{0.4pt} \, 0 \, \rule[.5ex]{1em}{0.4pt} & \left(\sigma_{w^{(4)}}^2\right)^T 
   \end{pmatrix}
\end{equation*}

\begin{equation*} 
    \sigma_{P_i}^2 = \sum_j \left( \sigma_{v_j}^2 W_{ji}^2 + v_j^2 (\sigma_W)_{ji}^2\right)
\end{equation*}

\begin{equation*}
    \frac{\partial P^{(group)}_i}{\partial P^{(l)}_i} =  \frac{\frac{\sum_{j\neq i}\prod_{k=1}^{n_k} \left(P_j^{(k)}\right)^{\frac{1}{n_k}}}{\prod_{k=1}^{n_k} \left(P_i^{(k)}\right)^{\frac{1}{n_k}}}}{\left[ 1+\frac{\sum_{j\neq i}\prod_{k=1}^{n_k} \left(P_j^{(k)}\right)^{\frac{1}{n_k}}}{\prod_{k=1}^{n_k} \left(P_i^{(k)}\right)^{\frac{1}{n_k}}}\right]^2}\frac{1}{n_k P_i^{(l)}} =  \frac{ P^{(group)}_i \left( 1-P^{(group)}_i\right)}{n_k P_i^{(l)}}
\end{equation*}

\begin{equation*}
    \frac{\partial P^{(group)}_i}{\partial P^{(l)}_q} =  -\frac{\frac{\prod_{k=1}^{n_k} \left(P_q^{(k)}\right)^{\frac{1}{n_k}}}{\prod_{k=1}^{n_k} \left(P_i^{(k)}\right)^{\frac{1}{n_k}}}}{\left[ 1+\frac{\sum_{j\neq i}\prod_{k=1}^{n_k} \left(P_j^{(k)}\right)^{\frac{1}{n_k}}}{\prod_{k=1}^{n_k} \left(P_i^{(k)}\right)^{\frac{1}{n_k}}}\right]^2}\frac{1}{n_k P_q^{(l)}}
    =  -\frac{\prod_{k=1}^{n_k} \left(\frac{P_q^{(k)}}{P_i^{(k)}}\right)^{\frac{1}{n_k}}}{\left[ 1+\frac{\sum_{j\neq i}\prod_{k=1}^{n_k} \left(P_j^{(k)}\right)^{\frac{1}{n_k}}}{\prod_{k=1}^{n_k} \left(P_i^{(k)}\right)^{\frac{1}{n_k}}}\right]^2}\frac{1}{n_k P_q^{(l)}} =  - \frac{P^{(group)}_i  P^{(group)}_q}{n_k P_q^{(l)}}
\end{equation*}

\begin{equation*} 
    \sigma_{P_i^{(group)}}^2 = \sum_l \left[\left(\frac{\partial P^{(group)}_i}{\partial P^{(l)}_i}
    \right)^2\sigma_{P^{(l)}_i}^2 + 
    \sum_{q\neq i}\left(\frac{\partial P^{(group)}_i}{\partial P^{(l)}_q}\right)^2\sigma_{P^{(l)}_q}^2\right]
\end{equation*}

\begin{equation*} 
   \sigma_{S}^2 =  \sum_{i=1}^{N_{ind}} \sigma_{P^{(group)}_i}^2 x_i^2 
\end{equation*}

```python
import numpy as np
import pandas as pd
from scipy.stats.mstats import gmean
import os
import glob
```

```python
RI=np.array([0,0,0,0.58,0.9,1.12,1.24,1.32,1.41,1.45,1.51]) #random index, il valore di m corrisponde all'indice dell'array
```

```python
def priority_vect(CoMa): # calcola i vettore di priorità della matrice di comparazione
    W_vect=gmean(CoMa, axis=1)
    Wstar=W_vect/np.sum(W_vect)
    return(Wstar)

def Con_Ind(MatrComp,PrioVect): #calcola gli indici di consisteza CI e CR dei confronti
    AWf=np.dot(MatrComp,PrioVect)
    AWfnorm=AWf/PrioVect
    #np.mean(AWfnorm)
    m=len(AWfnorm)
    CIf=(np.mean(AWfnorm)-m)/(m-1)
    if m<=2:
        CRf=0
    else:
        CRf=CIf/RI[m]
    return CIf, CRf   
```

```python
def sigma_error(matrix, priority):
    p_ratio = np.outer(priority, np.power(priority,-1))
    m_c = len(priority)
    return np.triu(np.square(np.log(np.divide(matrix, p_ratio)))).sum() *2 / ((m_c-1)*(m_c-2))
```

```python
def sigma_local_weight(weight, sigma_sq):
    m_c = weight.shape[0]
    weight_sq = np.square(weight)
    sigma_sq_local = (weight_sq.sum()-weight_sq) * sigma_sq * weight_sq * (1 - 1/(m_c**2))
    return sigma_sq_local 
```

```python
def block_matrix(vector_list):
    lenght_list = [len(x) for x in vector_list]
    n_col = sum(lenght_list)
    b_matrix = np.zeros((len(vector_list), n_col)) # costruisce la matrice a blocchi dei vettori di priorità per il calcolo del punteggio
    for i, v in enumerate(vector_list):
        if i==0:
            col_start = 0
            b_matrix[0,:v.shape[0]] = v
        else:
            col_start += lenght_list[i-1]
            b_matrix[i, col_start:(col_start + v.shape[0])] = v
    return b_matrix
```

```python
def global_weight_sing(file_):
    #print("Elaborazione del file {}".format(file_))
    excel_file = pd.ExcelFile(file_)
    C_matrix = excel_file.parse(sheet_name='PrefCri', header=None, usecols='C:F', skiprows=7).to_numpy()
    n_crit = C_matrix.shape[0]
    
    n_subcri= [5, 6, 6, 4]
    
    # Il numero di criteri e di sottocriteri e le matrici di comparazione vengono lette dal file excel "alberoAHP" che è situato nella stessa cartella dell'algoritmo
    SubC_list=[excel_file.parse(sheet_name=i+2, header=None, usecols=[j+2 for j in range(n_subcri[i])], skiprows=7).to_numpy() for i in range(n_crit)]    
    
    pri_v = priority_vect(C_matrix)
    
    # eleimino riga e colonna della matrice dei confronti a coppie relative 
    # all'indicatore 'costo totale utilizzo macchinari/costo totale progetto'
    # poiché questo indicatore viene scartato
    SubC_list[2] = np.delete(np.delete(SubC_list[2], 2, 0), 2, 1) 
    pri_w_list = [priority_vect(m) for m in SubC_list]
    
    W_matrix = block_matrix(pri_w_list) # costruisce la matrice a blocchi dei vettori di priorità per il calcolo del punteggio
    
    # peso globale        
    P = np.dot(pri_v, W_matrix) 
    
    sigma2 = sigma_error(C_matrix, pri_v)
    sigma_c2_list = [sigma_error(A, pri_w) for A, pri_w in zip (SubC_list, pri_w_list)]
    
    sigma_v2 = sigma_local_weight(pri_v, sigma2)
    sigma_w2_list = [sigma_local_weight(pri_w, sigma_c2) for pri_w, sigma_c2 in zip(pri_w_list, sigma_c2_list)]
    
    sigma_W2_block = block_matrix(sigma_w2_list)
                
    sigma_P2 = np.dot(sigma_v2, np.square(W_matrix))+ np.dot(np.square(pri_v), sigma_W2_block)
    return P, sigma_P2

```

```python
def consistency(file_):
    #print("Elaborazione del file {}".format(file_))
    excel_file = pd.ExcelFile(file_)
    C_matrix = excel_file.parse(sheet_name='PrefCri', header=None, usecols='C:F', skiprows=7).to_numpy()
    n_crit = C_matrix.shape[0]
    
    n_subcri= [5, 6, 6, 4]
    
    # Il numero di criteri e di sottocriteri e le matrici di comparazione vengono lette dal file excel "alberoAHP" che è situato nella stessa cartella dell'algoritmo
    SubC_list=[excel_file.parse(sheet_name=i+2, header=None, usecols=[j+2 for j in range(n_subcri[i])], skiprows=7).to_numpy() for i in range(n_crit)]    
    
    pri_v = priority_vect(C_matrix)
    
    # eleimino riga e colonna della matrice dei confronti a coppie relative 
    # all'indicatore 'costo totale utilizzo macchinari/costo totale progetto'
    # poiché questo indicatore viene scartato
    SubC_list[2] = np.delete(np.delete(SubC_list[2], 2, 0), 2, 1) 
    pri_w_list = [priority_vect(m) for m in SubC_list]
    
    print('Indici di consistenza per la matrice C:')
    print(Con_Ind(C_matrix,pri_v))
    for m_, v_ in zip(SubC_list, pri_w_list):
        print('Indici di consistenza per la matrice A:')
        print(Con_Ind(m_,v_))
    
```

```python
def ecdf(x):
    x = np.sort(x)
    def result(v):
        return np.searchsorted(x, v, side='right') / x.size
    return result

# cdf = ecdf(x)
# print(cdf(v))
```

```python
def one_minus_ecdf(x):
    x = np.sort(x)
    def result(v):
        return 1. - np.searchsorted(x, v, side='right') / x.size
    return result
```

```python
lista_ecdf = []
for c_ in indicatori_df.columns:
    if c_ in ('costoFornitori_su_costoTotale', 'oreTeam_su_costoTeam', 'brevetti_su_consulenti'):
        lista_ecdf.append(one_minus_ecdf(indicatori_df[c_].to_numpy()))
    else:
        lista_ecdf.append(ecdf(indicatori_df[c_].to_numpy()))
```
